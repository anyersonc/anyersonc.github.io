## Difussion

Difussion is modeled by Fick's first law of difussion, in which $\mathcal{F}$ represents the flux of the quantity, $D$ represents the diffusion coefficient that depends on the element and $\eta$ represents the concentration of the quantity. The derivative represents a gradient, a difference for making the process to happen, without a gradient, or in other words, if the gradient is zero, the total flux will be zero.

$$\mathcal{F}=-D\frac{\mathrm{d}\eta}{\mathrm{d}x} \tag{10.6}$$

In the case of electrons, we can model the flux of electrons by employing the same definition, however, it is important to address that it is a representation of the total movement of the electrons, and it is oversimplifying the short range of movement of electrons and its jumping between different levels.

For the flux of electrons we can rewrite the equation as follows:

$$\mathcal{F}_N=-D_N\frac{\mathrm{d}n}{\mathrm{d}x}  \tag{10.8}$$  

Whereas for the case of holes, we can rewrite the equation as follows:

$$\mathcal{F}_P=-D_P\frac{\mathrm{d}p}{\mathrm{d}x}  \tag{10.8a}$$


Then, if the flux of electrons is multiplied by the charge $-\mathrm{q}$, because the electrons posses negative charge, we have the **current density** *due to* **difussion** $J_{P\vert \mathrm{diff}}$:

$$J_{N\vert \mathrm{diff}}=-\mathrm{q}\mathcal{F}_N \tag{10.9}$$

And for the holes, which possesses positive charge $\mathrm{q}$ we have:

$$J_{P\vert \mathrm{diff}}=\mathrm{q}\mathcal{F}_P   \tag{10.9a}$$

By including equation $(10.8)$ into $(10.9)$:

$$J_{N\vert \mathrm{diff}}=\mathrm{q}D_N\frac{\mathrm{d}n}{\mathrm{d}x}  \tag{10.10}$$

By including equation $(10.8a)$ into $(10.9a)$:

$$J_{P\vert \mathrm{diff}}=-\mathrm{q}D_P\frac{\mathrm{d}p}{\mathrm{d}x}  \tag{10.11}$$



## The N-type and P-type semiconductors

## N-type semiconductors

The N-type semiconductor is the material obtained by employing a N-type doping as we recall from before. We could say that after the process, the material is now filled with atoms that possesses extra electrons. In an unbalanced semiconductor, more atoms of doping material are on one side of the semiconductor, creating a movement or flux of electrons, which will move to try to balance the charge. By doing that, it generates a current density caused by diffusion and creating an electric field.

$$ J_N = q \mu_{n}n \mathcal{E}_x + q D_N\frac{dn}{dx} \tag{10.15a}$$

The p-type semiconductor has a similiar function as the N-type, the main difference is the holes doping of the material. The same process apply, as the holes are concentrated in one side, the material will balance this charge producing a holes flux density, generating a current density by diffusion, and an electric field.

The following equations, $(11.1a)$ and $(11.1b)$ presents the current density in a $pn$-junction, as we can notice, the current generated by the electron or holes drift is balanced by the electron or hole difussion:

$$ J_N = J_{N \vert drift} + J_{N \vert diff} \tag{11.1a}$$

$$ J_P = J_{P \vert drift} + J_{P \vert diff} \tag{11.1b}$$

Additionally, we can write the charge concentration inside the device as the sum of the contributions due to each entities with charge, in this case: electrons, holes, ionized donors and ionized acceptors, as expressed in the equation $(11.3)$ note that the units will be charge/$\mathscr{cm}^2$:

$$ \frac{charge}{cm^3} = qp-qn+qN_{D}^{+}-qN_{A}^{-} \tag{11.3} $$

From the previous, note that $N_{D}^{+}$ and $N_{A}^{-}$ are the concentrations of ionized donors and acceptors respectively. Now, we have to discuss a new concept, Silicon at room temperature has sufficient energy for nearly all the dopants to ionize, and if we apply the total ionization assumption we have the equation $(11.4)$ and $(11.5)$.

$$ N_{D}^{+}=N_{D} \tag{11.4}$$

$$ N_{A}^{-}=N_{A} \tag{11.5}$$

Now, taking into account the previous, we can substitute $(11.4)$ and $(11.5)$ into $(11.3)$ we arrive to the equation $(11.6)$ which expresses the total charge density in the device:

$$\rho = q(p-n + N_{D}-N_{A}) \tag{11.6}$$

Now, taking into account the charge density, it is possible to establish the electric field in the junction using the Gauss's law presented in the equation $(11.7)$.

$$ \nabla \cdot \mathscr{E} = \frac{\rho}{\epsilon_r \epsilon_0}  \tag{11.7} $$

If we neglect the effects of the edges in our ideal junction, the electric field has only the $x$ dimension component, so, the equation $(11.7)$ could be expressed as $(11.8)$ where the $\epsilon_r$ is the relative permittivity of the semiconductor and $\epsilon_0$ is the permittivity of free space.

$$ \frac{\mathrm{d}\mathscr{E}}{\mathrm{d}x} = \frac{\rho}{\epsilon_r \epsilon_0}  \tag{11.8} $$

The equation $(11.8)$ shows that the slope of the electric field is proportional to the charge density $\rho$, and if we take into account that the charge density is approximately zero outside the depletion region, the electric field there must be constant, which in fact, is stated that the electric field in a small distance from the outside edges of the depletion region is exactly zero. Then, if we define the $-x_p$ and $x_n$ to be the left and right sides of the depletion region, the electric field of the region can be found of integrating $(11.8)$, and if we are interested in any point, called $f$ then our integral will be defined bewteen $-x_p$ to $f$ as in the equation $(11.9)$, which curve is presented in the Figure (c).

$$ \mathscr{E}_x ( f )= \int_{-x_p}^{f} \frac{\rho}{\epsilon_r\epsilon_0} \mathrm{d}x \tag{11.9} $$

Hence, the change in voltage with increasing $x$ can be calculated in a path as $(11.10)$

$$ V ( f )- V(i)= -\int_{i}^{f} \mathscr{E}_x \mathrm{d}x \tag{11.10} $$

Now, recalling that we set the electric field zero outside the depletion region, then we chose to set the voltage on the $p$-side of the depletion region to zero. and because we are integrating from $-x_p$, the previous equations then it is simplify as appears in the equation $(11.11)$ whose graph appears in the Figure (d).

$$ V ( f )= -\int_{-x_p}^{f} \mathscr{E}_x \mathrm{d}x \tag{11.11}$$

For instance, the potencial change acrros the depletion region under equilibrium is known as **built-in potential** $V_{\mathrm{bi}}$ which can be calculated by employing the equation $(11.12)$.

$$ V_{\mathrm{bi}} = V ( x_n)- V(-x_p)=-\int_{-x_p}^{x_n} \mathscr{E}_x \mathrm{d}x \tag{11.12}$$

Now, we can work for deriving an expression for the **built-in potential** $V_{\mathrm{bi}}$ that uses the concentrations, temperature and physical constants. For doing that, we recall the equation $(11.13)$ which represents equilibrium, so the currents must sum zero.

$$ J_{N\vert drift} + J_{N\vert diff} \tag{11.13}$$

Then, expanding the current components using equation $(10.15a)$ we found the equation $(11.14)$

$$ q\mu_nn\mathscr{E}_x + qD_n\frac{\mathrm{d}n}{\mathrm{d}x} \tag{11.14}$$

If we solve for $\mathscr{E}_x$ it results in the equation $(11.15)$

$$ \mathscr{E}_x = \frac{-D_n}{\mu_n}\frac{1}{n}\frac{\mathrm{d}n}{\mathrm{d}x}  \tag{11.15}$$

If we employ the Enstein's relation given by $D_n / \mu_n = k T / q$, and replace $D_n$ we find an expression that depends on physical constants, temperature and electron concentration shown in $(11.16)$

$$ \mathscr{E}_x = \frac{-kT}{q}\frac{1}{n}\frac{\mathrm{d}n}{\mathrm{d}x}  \tag{11.16}$$

Now, using this expression into the equation $(11.12)$ it yields to the equation $(11.17)$

$$  V_{\mathrm{bi}} = \int_{-x_p}^{x_n} \frac{1}{n}\frac{\mathrm{d}n}{\mathrm{d}x} \mathrm{d}x \tag{11.17}$$

Yet, if we use a change of variables of integration for solving the integral, we can rewrite it as appears in the equation $(11.18)$

$$  V_{\mathrm{bi}} = \frac{kT}{q}\int_{n(-x_p)}^{n(x_n)} \frac{1}{n}\mathrm{d}x  \tag{11.18}$$

Which we can solve obtaining the equation $(11.19)$

$$  V_{\mathrm{bi}} = \frac{kT}{q}\ln{\frac{n(n_x)}{n(n(-x_p)}}  \tag{11.19}$$

Now, we need to include the total ionization assumption, and by doing so we can see that $n$ at the point $x_n$ will lead to the equation $(11.20)$ and for the $p$ at $-x_p$ will lead to the equation $(11.21)$

$$ n(x_n)=N_D \tag{11.20}$$

$$ p(-x_p)=N_A \tag{11.21}$$

Then, if we use the $np$ product rule, we can find the equation $(11.22)$ 

$$ n(-x_p)=\frac{n_{i}^{2}}{p(-x_p)}=\frac{n_{i}^{2}}{N_A} \tag{11.22}$$

And, finally, substituting these results back into $(11.19)$ we arrived to the equation $(11.23)$ which is an expression of $V_{\mathrm{bi}}$ as a function of the device's doping concentrations, temperature and physical constants.

$$  V_{\mathrm{bi}}= \frac{kT}{q} \ln{ \frac{N_AN_D}{n_{i}^{2}}} \tag{11.23} $$


## Depletion Width
In general the charge density within the depletion region is not known ahead of time because it depends on the carrier concentrations p and n which in turn depend on the electric potential.
In some cases, we may be able to write expressions for p and n as a function of x which allow for a closed form solution to (11.25), but more often we apply a simple approximation to the charge carriers known as the depletion approximation.

$$ \frac{\mathrm{d}\mathscr{E}_x}{\mathrm{d}x} = \frac{\rho}{\epsilon_r \epsilon_0} = \frac{q}{\epsilon_r\epsilon_0 (p-n+N_D-N_A)}  \tag{11.25}$$

Under the depletion region approximation, the total charge density $\rho$  of a pn step junction takes on a squared off form relative to its exact value. Stated formally:

$$ \rho = \Biggl\{\begin{array}{rc}
0 & \mathrm{if} \space x \leq -x_p \mathrm{or} \space x \geq x_n
\\ -qN_A & \mathrm{if} \space -x_p \leq x \leq 0 \\
qN_D & \mathrm{if} \space 0 \leq x \leq x_n
\end{array} $$

To completely specify the depletion approximation, we need to establish  boundaries on each side of the junction. We will begin by first working out the widths $x_p$ and $x_n$ relative to each other. The width of the depletion region on each side of the junction is dependent on the relative doping level between the two sides such that the p-side and n-side widths will only be equal. The excess negative charge on the p-side of the junction must precisely balance the excess positive charge on the n-side of the junction.

$$ Q_P^{-} = Q_n^{+} \tag{11.27}$$

Due to the depletion approximation, these excess charges can be expressed as the charge density multiplied by the volume of the depletion region on each side of the junction.

$$ qN_AV_\mathrm{P}=qN_DV_n \tag{11.28}$$

The volume of the depletion region on a given side of the junction is equal to the depletion width of that side times the cross-sectional area of the device, which is assumed to be constant.

$$ \begin{array}{ll} V_\mathrm{P}=x_\mathrm{P}A, & V_\mathrm{n}=x_\mathrm{n}A \end{array} \tag{11.29}$$

 By dropping the area from each side of the equation, the charge balance between each side can be expressed as

$$ qN_Ax_\mathrm{P} = qN_Dx_\mathrm{n} \tag{11.30}$$

Finally we can relate the relative widths of the depletion region on each side of the junction by solving for $x_n$ in terms of $x_p$ which results in

$$ x_\mathrm{n} = \frac{N_A}{N_D}x_\mathrm{P} \tag{11.31}$$

At this point we have established one equation relating xp and xn (11.30), but because we have two unknowns, we require a second independent equation relating them in order to locate both endpoints of the depletion region. A second equation relating $x_p$ and $x_n$ can be found by considering the electric potential within the junction. In order to determine the electric potential, we will first need to establish the electric field, which can be calculated from the charge distribution. We begin by substituting the charge density we found using the depletion approximation into the one dimensional version of Gauss’s law in differential form from (11.25).

$$ \frac{\mathrm{d}\mathscr{E}_x}{\mathrm{d}x} = \frac{\rho}{\epsilon_r \epsilon_0} = \Biggl\{\begin{array}{rc}
0 & \mathrm{if} \space x \leq -x_p \mathrm{or} \space x \geq x_n
\\ \dfrac{-qN_A}{\epsilon_r\epsilon_0} & \mathrm{if} \space -x_p \leq x \leq 0 \\
\mathrm{qN_D}{\epsilon_r\epsilon_0} & \mathrm{if}\space  0  \leq x \leq x_n
\end{array} $$

The electric field is zero and constant outside of the depletion region. Thus the electric field must be zero at the depletion region edges -$x_p$ and $x_n$. Then for the region between $-x_p$ and $0$ we have

$$ \frac{\mathrm{d}\mathscr{E}_x}{\mathrm{d}x} = -\frac{qN_A}{\epsilon_r\epsilon_0} \tag{11.34}$$

We solve for $\mathscr{E}_x(x)$ in the interval by separating variables and integrating each side. The lower bounds are fixed at $\mathscr{E}_x, = -x_p$ while the upper bounds vary through the extent of the interval.

$$ \int_{0}^{\mathscr{E}_x(x)} \mathrm{d}\dot{\mathscr{E}_x} = \int_{-x_p}^{x} \frac{-qN_A}{\epsilon_r\epsilon_0} \mathrm{d}\dot{x} \tag{11.35}$$

Completing the integration yields

$$\begin{array}{ll} \mathscr{E}_x(x) = \dfrac{-qN_A}{\epsilon_r\epsilon_0}(x +x_\mathrm{p}), & \mathrm{for} \space -x_\mathrm{P} \leq x \leq 0 \end{array} \tag{11.36}$$

 Next, for the region between 0 and xn we have

$$ \frac{\mathrm{d}\mathscr{E}_x}{\mathrm{d}x} = \frac{qN_D}{\epsilon_r\epsilon_0} \tag{11.37}$$

Again separating variables and integrating each side but this time fixing the upper bounds at $\mathscr{E}_x=0, x=x_n$ and letting the lower bounds vary throughout the interval we have

$$ \int_{\mathscr{E}_x(x)}^{0} \mathrm{d}\dot{\mathscr{E}_x} = \int_{x}^{x_n} \frac{-qN_D}{\epsilon_r\epsilon_0} \mathrm{d}\dot{x} \tag{11.38}$$

Completing the integration yields

$$\begin{array}{ll} \mathscr{E}_x(x) = \dfrac{-qN_D}{\epsilon_r\epsilon_0}(x_\mathrm{n}-x), & \mathrm{for} \space 0 \leq x \leq x_\mathrm{n} \end{array} \tag{11.39}$$

From the resulting expressions in equations (11.36) and (11.39) we see that the electric field in the depletion region is always negative and that it decreases linearly from $x = -x_p$ until $x=0$ and then increases linearly until $x=x_n$. If we assume that the electric field at the inflection point $(x = 0)$ is continuous, this means that (11.36) and (11.39) must be equal at x = 0. Hence,

$$\frac{-qN_A}{\epsilon_r\epsilon_0}x_\mathrm{P} = \frac{-qN_D}{\epsilon_r\epsilon_0}x_\mathrm{n} \tag{11.40}$$

which can be simplified to

$$qN_Ax_\mathrm{P} = qN_Dx_\mathrm{n} \tag{11.41}$$

Now we define the electric potential in the $x \leq -x_p$ region to be zero volts. Due to the piecewise linear nature of our solution for $\mathscr{E}_x$, we break the integral of (11.10) into intervals. For the interval on the $p$-side of the junction, we let the upper bound vary through the extent of the interval resulting in

$$ V ( x ) - V ( -x_\mathrm{P} ) = V ( x ) - 0 = \int_{-x_\mathrm{P}}^{x} \dfrac{qN_A}{\epsilon_r\epsilon_0} ( x_\mathrm{P} + \dot{x}) \mathrm{d}x \tag{11.42}$$

which upon completing the integration provides

$$\begin{array}{ll} V ( x ) = \dfrac{qN_A}{2\epsilon_r\epsilon_0}(x_\mathrm{p}+x)^2, & \mathrm{for} \space -x_\mathrm{P} \leq x \leq 0 \end{array} \tag{11.43}$$

For the interval on the $n$-side of the junction we vary the lower bound throughout the interval such that

$$ V ( x_\mathrm{n} ) - V ( x ) = V_{\mathrm{bi}} - V ( x ) = \int_{-x_\mathrm{P}}^{x} \dfrac{qN_D}{\epsilon_r\epsilon_0} ( x_\mathrm{n} - \dot{x}) \mathrm{d}x \tag{11.44}$$

which upon completing the integration provides

$$\begin{array}{ll} V ( x ) =  V_{\mathrm{bi}} - \dfrac{qN_D}{2\epsilon_r\epsilon_0}(x_\mathrm{n}-x)^2, & \mathrm{for} \space 0 \leq x \leq x_\mathrm{n} \end{array} \tag{11.45}$$

Since the voltage is continuous we can set (11.43) equal to (11.45) at $x=0$ to arrive at a second equation relating $x_p$ and $x_n$.

$$ \frac{qN_A}{2\epsilon_r\epsilon_0} x_\mathrm{P}^2 = V_\mathrm{bi}-\frac{qN_D}{2\epsilon_r\epsilon_0}x_n^2 \tag{11.46}$$

Using (11.31) and (11.46) we can now express the $p-$side and $n-$side depletion widths ($x_p$ and $x_n$) in terms of dopant levels.We begin by substituting the solution for $x_n$ in terms of $x_p$ from (11.31) into (11.46) resulting in

$$ \frac{qN_A}{2\epsilon_r\epsilon_0} x_\mathrm{P}^2 = V_\mathrm{bi}-\frac{qN_D}{2\epsilon_r\epsilon_0}\frac{N_A^2}{N_D^2}x_\mathrm{P}^2 \tag{11.47}$$

With a bit more algebra we can solve for $x_p$ to find

$$ x_\mathrm{P} = \left[ \frac{2\epsilon_r\epsilon_0}{q} \left( \frac{N_D}{N_A (N_A + N_D )} \right) V_\mathrm{bi} \right]^{1/2} \tag{11.48}$$

Then by using (11.31) again we can also solve for $x_n$ to find

$$ x_\mathrm{n} = \frac{N_A}{N_D}x_\mathrm{P} = \left[ \frac{2\epsilon_r\epsilon_0}{q} \left( \frac{N_A}{N_D (N_A + N_D )} \right) V_\mathrm{bi} \right]^{1/2} \tag{11.49}$$

By expressing the depletion width on each side of the junction in terms of the dopant levels and physical parameters in (11.48) and(11.49),we have completely specified the depletion approximation.

The total depletion region width, $W$ can be defined as the sum of the $p$-side and $n$-side depletion region widths.

$$ W = x_\mathrm{P} + x_\mathrm{n} \tag{11.50}$$

By choosing a common denominator in the expressions for $x_p$ and $x_n$, and combining like terms, (11.50) can be written as

$$ W = \left[ \frac{2\epsilon_r\epsilon_0}{q}\right]^{1/2} \left( \frac{N_D+N_A}{[N_AN_D (N_A + N_D )]^{1/2}} \right) V_\mathrm{bi}^{1/2}  \tag{11.51}$$

Then by cancelling a factor of $\sqrt{N_A + N_D}$ and grouping all the terms under same square root we arrive at

$$ W = \left[ \frac{2\epsilon_r\epsilon_0}{q} \left( \frac{N_D+N_A}{N_AN_D} \right) V_\mathrm{bi} \right]^{1/2}  \tag{11.52}$$

## Band Bending
To begin we relate voltage to energy by noting that the electric potential $V$ is equivalent to the electric potential energy (P.E.) per unit of charge at a given point in an electric field. Hence,

$$ V= \frac{P.E.}{q} \tag{11.53}$$

where $q$ is the elementary charge $(1.6 \times 10^{-19} C)$. For an electron with charge $-q$ we can write its electrical potential energy in terms of the electric potential as

$$ P.E. = -qV \tag{11.54} $$

## $I$-$V$ Characteristics

The difference in potential between the $n$-side and the $p$-side of the junction is given by $V_{bi} - V_A$ from which we see that a positive applied voltage reduces the potential difference between the two sides whereas a negative applied voltage increases it. A positive applied voltage decreases the depletion width whereas a negative applied voltage increases it. Equations (11.48), (11.49), and (11.52) for depletion widths in an isolated $pn$ junction can be generalized to account for the applied voltage by replacing $V_{bi}$ in each equation with a $V_{bi} - V_A$ such that

$$ x_p = \left[ \frac{2\epsilon_r\epsilon_0}{q} \frac{N_D}{N_A(N_A + N_D)}(Vbi - V_A)\right]^{1/2} \tag{11.55} ,$$

$$ x_n = \left[ \frac{2\epsilon_r\epsilon_0}{q} \frac{N_A}{N_D(N_A + N_D)}(Vbi - V_A)\right]^{1/2} \tag{11.56} ,$$

and

$$ W = \left[ \frac{2\epsilon_r\epsilon_0}{q} \left (\frac{N_A + N_D}{N_AN_D}\right )(Vbi - V_A)\right]^{1/2} \tag{11.57}$$

Mathematically, the complite $i-v$ characteristic curve for an ideal $pn$ junction diode is given by

$$ I=I_0 (e^{\frac{qV_A}{kT}}-1) \tag{11.58}$$
