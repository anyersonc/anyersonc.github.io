# Worked example


<p>Now go to <a href="https://nanohub.org/tools/mosfet" target="_blank" rel="noopener noreferrer">NanoHub MOSFet</a> to access the simulation tool.</p> 

For accessing, recall the $pn$ junctions worked example for entering into Nanohub simulations.

The objective of this activity is to find the effect on the $I_\mathrm{D}-V_\mathrm{D}$ curve characteristic caused by the temperature. Follow the next steps to get the results.

## Input Parameters

In this simulation we only are going to modify the temperature with the standard value to observe the isolated effect of the temperature on the device behavior.

$T_1= 200 \mathrm{K}$

$T_2= 400 \mathrm{K}$

## Step by Step procedure

### (a) Enter to Nanohub

Enter into NanoHub and clic on "Launch tool."

<center>
<img src="https://raw.githubusercontent.com/anyersonc/anyersonc.github.io/main/NanoMos.png" alt='Image nanohub'>
</center>

### (b) Find the temperature parameter

In the tool, clic on the "Model" tab in the part above. It is the middle tab in the tool.

<center>
<img src="https://raw.githubusercontent.com/anyersonc/anyersonc.github.io/main/Mos1.png" alt='Image nanohub'>
</center>

### (c) Check the current temperature and change it

Now in the Model tab, we find the Temperature option. You should clic on that for accessing the temperature change. Note that the temperature is in Kelvin, $\mathrm{K}$, units.

<center>
<img src="https://raw.githubusercontent.com/anyersonc/anyersonc.github.io/main/Mos2.png" alt='Image nanohub'>
</center>

### (d) Choose the first temperature $T_1$

Select the option or write the option of the temperature for the simulation, remember that for this case $T_1=200\mathrm{K}$

<center>
<img src="https://raw.githubusercontent.com/anyersonc/anyersonc.github.io/main/Mos3.png" alt='Image nanohub'>
</center>

### (e) Simulate the first temperature $T_1$

Once selected, clic on simulate for obtaining the curves.

<center>
<img src="https://raw.githubusercontent.com/anyersonc/anyersonc.github.io/main/Mos4.png" alt='Image nanohub'>
</center>

### (f) Check if the simulation produces coherent curves

The simulation must provide two curves, one red and one blue, that must represent the behavior expected of a MOSFET with a gate voltage $V_\mathrm{G} \gt V_\mathrm{T}$. In the red curve we will find the case of $V_\mathrm{D} \lt V_\mathrm{D_\mathrm{sat}}$ and in the blue curve we will find the case for $V_\mathrm{D} \gt V_\mathrm{D_\mathrm{sat}}$.


<center>
<img src="https://raw.githubusercontent.com/anyersonc/anyersonc.github.io/main/Mos5.png" alt='Image nanohub'>
</center>

### (g) Choose the second temperature $T_2$

For the second case, we will simulate in the same simulation, we will not terminate or restart the tool. Now, choosing for the second temperature $T_2=400\mathrm{K}$. 

<center>
<img src="https://raw.githubusercontent.com/anyersonc/anyersonc.github.io/main/Mos6.png" alt='Image nanohub'>
</center>

### (h) Check if the simulation produces coherent curves

The simulation must provide two curves, one red and one blue, that must represent the behavior expected of a MOSFET with a gate voltage $V_\mathrm{G} \gt V_\mathrm{T}$. In the red curve we will find the case of $V_\mathrm{D} \lt V_\mathrm{D_\mathrm{sat}}$ and in the blue curve we will find the case for $V_\mathrm{D} \gt V_\mathrm{D_\mathrm{sat}}$.

<center>

### (i) Overlap the curves for comparison and analysis

Now with the option in the right inferior side, clic on "All" for plotting the two results at the same time. Now it is posible to compare the two cases and produce an analysis about the effect of the temperature on the behavior of a MOSFET.
