## Difussion

Difussion is modeled by Fick's first law of difussion, in which $\mathcal{F}$ represents the flux of the quantity, $D$ represents the diffusion coefficient that depends on the element and $\eta$ represents the concentration of the quantity. The derivative represents a gradient, a difference for making the process to happen, without a gradient, or in other words, if the gradient is zero, the total flux will be zero.

$$\mathcal{F}=-D\frac{\mathrm{d}\eta}{\mathrm{d}x} \tag{1}$$

In the case of electrons, we can model the flux of electrons by employing the same definition, however, it is important to address that it is a representation of the total movement of the electrons, and it is oversimplifying the short range of movement of electrons and its jumping between different levels.

For the flux of electrons we can rewrite the equation as follows:

$$\mathcal{F}_N=-D_N\frac{\mathrm{d}n}{\mathrm{d}x}  \tag{2}$$  

Whereas for the case of holes, we can rewrite the equation as follows:

$$\mathcal{F}_P=-D_P\frac{\mathrm{d}p}{\mathrm{d}x}  \tag{3}$$


Then, if the flux of electrons is multiplied by the charge $-\mathrm{q}$, because the electrons posses negative charge, we have the **current density** *due to* **difussion** $J_{P\vert \mathrm{diff}}$:

$$J_{N\vert \mathrm{diff}}=-\mathrm{q}\mathcal{F}_N \tag{4}$$

And for the holes, which possesses positive charge $\mathrm{q}$ we have:

$$J_{P\vert \mathrm{diff}}=\mathrm{q}\mathcal{F}_P   \tag{5}$$

By including equation $(2)$ into $(4)$:

$$J_{N\vert \mathrm{diff}}=\mathrm{q}D_N\frac{\mathrm{d}n}{\mathrm{d}x}  \tag{6}$$

By including equation $(3)$ into $(5)$:

$$J_{P\vert \mathrm{diff}}=-\mathrm{q}D_P\frac{\mathrm{d}p}{\mathrm{d}x}  \tag{6}$$

## Depletion Width
In general the charge density within the depletion region is not known ahead of time because it depends on the carrier concentrations p and n which in turn depend on the electric potential.
In some cases, we may be able to write expressions for p and n as a function of x which allow for a closed form solution to (11.25), but more often we apply a simple approximation to the charge carriers known as the depletion approximation.

Equation 11.25

$$ \frac{\mathrm{d}\mathscr{E}_x}{\mathrm{d}x} = \frac{\rho}{\epsilon_r \epsilon_0} = \frac{q}{\epsilon_r\epsilon_0 (p-n+N_D-N_A)}  \tag{11.25}$$

Under the depletion region approximation, the total charge density $\rho$  of a pn step junction takes on a squared off form relative to its exact value. Stated formally:

Equation 11.26

$$ \rho = \Biggl\{\begin{array}{rc}
0 & \mathrm{if} \space x \leq -x_p \mathrm{or} \space x \geq x_n
\\ -qN_A & \mathrm{if} \space -x_p \leq x \leq 0 \\
qN_D & \mathrm{if} \space 0 \leq x \leq x_n
\end{array} $$

To completely specify the depletion approximation, we need to establish  boundaries on each side of the junction. We will begin by first working out the widths $x_p$ and $x_n$ relative to each other. The width of the depletion region on each side of the junction is dependent on the relative doping level between the two sides such that the p-side and n-side widths will only be equal. The excess negative charge on the p-side of the junction must precisely balance the excess positive charge on the n-side of the junction.

Equation 11.27

$$ Q_P^{-} = Q_n^{+} \tag{11.27}$$

Due to the depletion approximation, these excess charges can be expressed as the charge density multiplied by the volume of the depletion region on each side of the junction.

Equation 11.28

$$ qN_AV_\mathrm{P}=qN_DV_n \tag{11.28}$$

The volume of the depletion region on a given side of the junction is equal to the depletion width of that side times the cross-sectional area of the device, which is assumed to be constant.

Equation 11.29

$$ \begin{array}{ll} V_\mathrm{P}=x_\mathrm{P}A, & V_\mathrm{n}=x_\mathrm{n}A \end{array} \tag{11.29}$$

 By dropping the area from each side of the equation, the charge balance between each side can be expressed as

Equation 11.30

$$ qN_Ax_\mathrm{P} = qN_Dx_\mathrm{n} \tag{11.30}$$

Finally we can relate the relative widths of the depletion region on each side of the junction by solving for $x_n$ in terms of $x_p$ which results in

Equation 11.31

$$ x_\mathrm{n} = \frac{N_A}{N_D}x_\mathrm{P} \tag{11.31}$$

At this point we have established one equation relating xp and xn (11.30), but because we have two unknowns, we require a second independent equation relating them in order to locate both endpoints of the depletion region. A second equation relating $x_p$ and $x_n$ can be found by considering the electric potential within the junction. In order to determine the electric potential, we will first need to establish the electric field, which can be calculated from the charge distribution. We begin by substituting the charge density we found using the depletion approximation into the one dimensional version of Gauss’s law in differential form from (11.25).

Equation 11.33

$$ \frac{\mathrm{d}\mathscr{E}_x}{\mathrm{d}x} = \frac{\rho}{\epsilon_r \epsilon_0} = \Biggl\{\begin{array}{rc}
0 & \mathrm{if} \space x \leq -x_p \mathrm{or} \space x \geq x_n
\\ \dfrac{-qN_A}{\epsilon_r\epsilon_0} & \mathrm{if} \space -x_p \leq x \leq 0 \\
\mathrm{qN_D}{\epsilon_r\epsilon_0} & \mathrm{if}\space  0  \leq x \leq x_n
\end{array} $$

The electric field is zero and constant outside of the depletion region. Thus the electric field must be zero at the depletion region edges -$x_p$ and $x_n$. Then for the region between $-x_p$ and $0$ we have

Equation 11.34

$$ \frac{\mathrm{d}\mathscr{E}_x}{\mathrm{d}x} = -\frac{qN_A}{\epsilon_r\epsilon_0} \tag{11.34}$$

We solve for $\mathscr{E}_x(x)$ in the interval by separating variables and integrating each side. The lower bounds are fixed at $\mathscr{E}_x, = -x_p$ while the upper bounds vary through the extent of the interval.

Equation 11.35

$$ \int_{0}^{\mathscr{E}_x(x)} \mathrm{d}\dot{\mathscr{E}_x} = \int_{-x_p}^{x} \frac{-qN_A}{\epsilon_r\epsilon_0} \mathrm{d}\dot{x} \tag{11.35}$$

Completing the integration yields

Equation 11.36

$$\begin{array}{ll} \mathscr{E}_x(x) = \dfrac{-qN_A}{\epsilon_r\epsilon_0}(x +x_\mathrm{p}), & \mathrm{for} \space -x_\mathrm{P} \leq x \leq 0 \end{array} \tag{11.36}$$

 Next, for the region between 0 and xn we have

Equation 11.37

$$ \frac{\mathrm{d}\mathscr{E}_x}{\mathrm{d}x} = \frac{qN_D}{\epsilon_r\epsilon_0} \tag{11.37}$$

Again separating variables and integrating each side but this time fixing the upper bounds at $\mathscr{E}_x=0, x=x_n$ and letting the lower bounds vary throughout the interval we have

Equation 11.38

$$ \int_{\mathscr{E}_x(x)}^{0} \mathrm{d}\dot{\mathscr{E}_x} = \int_{x}^{x_n} \frac{-qN_D}{\epsilon_r\epsilon_0} \mathrm{d}\dot{x} \tag{11.38}$$

Completing the integration yields

Equation 11.39

$$\begin{array}{ll} \mathscr{E}_x(x) = \dfrac{-qN_D}{\epsilon_r\epsilon_0}(x_\mathrm{n}-x), & \mathrm{for} \space 0 \leq x \leq x_\mathrm{n} \end{array} \tag{11.39}$$

From the resulting expressions in equations (11.36) and (11.39) we see that the electric field in the depletion region is always negative and that it decreases linearly from $x = -x_p$ until $x=0$ and then increases linearly until $x=x_n$. If we assume that the electric field at the inflection point $(x = 0)$ is continuous, this means that (11.36) and (11.39) must be equal at x = 0. Hence,

Equation 11.40

$$\frac{-qN_A}{\epsilon_r\epsilon_0}x_\mathrm{P} = \frac{-qN_D}{\epsilon_r\epsilon_0}x_\mathrm{n} \tag{11.40}$$

which can be simplified to

Equation 11.41

$$qN_Ax_\mathrm{P} = qN_Dx_\mathrm{n} \tag{11.41}$$

...

Equation 11.42

$$ V ( x ) - V ( -x_\mathrm{P} ) = V ( x ) - 0 = \int_{-x_\mathrm{P}}^{x} \dfrac{qN_A}{\epsilon_r\epsilon_0} ( x_\mathrm{P} + \dot{x}) \mathrm{d}x \tag{11.42}$$

Equation 11.43

$$\begin{array}{ll} V ( x ) = \dfrac{qN_A}{2\epsilon_r\epsilon_0}(x_\mathrm{p}+x)^2, & \mathrm{for} \space -x_\mathrm{P} \leq x \leq 0 \end{array} \tag{11.43}$$

Equation 11.44

$$ V ( x_\mathrm{n} ) - V ( x ) = V_{\mathrm{bi}} - V ( x ) = \int_{-x_\mathrm{P}}^{x} \dfrac{qN_D}{\epsilon_r\epsilon_0} ( x_\mathrm{n} - \dot{x}) \mathrm{d}x \tag{11.44}$$

$$ \frac{qN_A}{2\epsilon_r\epsilon_0} x_\mathrm{P}^2 = V_\mathrm{bi}-\frac{qN_D}{2\epsilon_r\epsilon_0}x_n^2 \tag{11.46}$$


Equation 11.47

$$ \frac{qN_A}{2\epsilon_r\epsilon_0} x_\mathrm{P}^2 = V_\mathrm{bi}-\frac{qN_D}{2\epsilon_r\epsilon_0}\frac{N_A^2}{N_D^2}x_\mathrm{P}^2 \tag{11.47}$$

$$ x_\mathrm{P} = \left[ \frac{2\epsilon_r\epsilon_0}{q} \left( \frac{N_D}{N_A (N_A + N_D )} \right) V_\mathrm{bi} \right]^{1/2} \tag{11.48}$$
